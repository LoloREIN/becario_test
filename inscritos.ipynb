```python
import pandas as pd
```

```python
df = pd.read_csv("Inscritos Inteligencia Mecatronica.csv")
df.head(100)
```

```python
df.describe()
```

```python
df_filter = df[["CELULAR","E-MAIL","PATERNO","MATERNO","NOMBRE","NOMBRE PREFERIDO","NOMBRE COMPLETO","CARRERA","CORREO2"]].applymap(lambda x: x.lower() if isinstance(x, str) else x)
df_filter.shape
```

```python
df_conv = pd.read_csv("Conivio Ingeniería Responses.csv")
df_conv.head()
```

```python
df_conv_f = df_conv[df_conv["¿De qué Ingeniería eres?"].isin(("Ing. Mecatrónica","Ing. en Inteligencia de Datos y Ciberseguridad"))]
df_conv_f = df_conv_f.reset_index(drop=True)
df_conv_f
```

```python
df_conv_f_select = (df_conv_f[["Teléfono Celular (para mandar recordatorio)","Correo","Apellido Paterno","Apellido Materno","Nombre","¿De qué Ingeniería eres?"]]).applymap(lambda x: x.lower() if isinstance(x, str) else x)
df_conv_f_select.shape
```

```python
df_join_conv_inscritos = df_filter[~df_filter['E-MAIL'].isin(df_conv_f_select['Correo'])]
df_join_conv_inscritos.shape
```

```python
df_join_conv_inscritos["CORREO2"] = df_join_conv_inscritos["CORREO2"].fillna('')
df_join_conv_inscritos.head()
df_join_conv_inscritos = df_join_conv_inscritos[~df_join_conv_inscritos["CORREO2"].isin(df_conv_f_select['Correo'])]
df_join_conv_inscritos.shape
```

```python
df_join_conv_inscritos["CELULAR"] = df_join_conv_inscritos["CELULAR"].str.replace(" ", "", regex=False)
df_join_conv_inscritos["CELULAR"] = df_join_conv_inscritos["CELULAR"].str.replace("-","")
df_join_conv_inscritos["CELULAR"] = df_join_conv_inscritos["CELULAR"].str.replace("+","")
df_join_conv_inscritos["CELULAR"] = df_join_conv_inscritos["CELULAR"].str[-10:]
df_join_conv_inscritos
```

```python
df_join_conv_inscritos.to_excel('output.xlsx', index=False, sheet_name='Datos')
```
